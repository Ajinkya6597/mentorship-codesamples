Seaborn is a popular data visualization library built on top of Matplotlib. It provides a high-level interface for creating beautiful and informative statistical graphics. Here are a few examples of how you can use Seaborn to visualize data:

**Scatter plot:**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
df = sns.load_dataset('tips')

# Create a scatter plot of total_bill vs tip
sns.scatterplot(x='total_bill', y='tip', data=df)

# Show the plot
plt.show()
```

```python
# load the example iris dataset
iris = sns.load_dataset("iris")

# create a scatter plot of petal length vs petal width
sns.scatterplot(x="petal_length", y="petal_width", hue="species", data=iris)

# show the plot
plt.show()
```

```python
# load the example iris dataset
iris = sns.load_dataset("iris")

sns.scatterplot(x="petal_length", 
                y="petal_width", 
                hue="species", 
                palette="YlGnBu", 
                data=iris)

# show the plot
plt.show()
```

**Line Plot:** Use `lineplot()` function to create a line plot. It can be useful for showing trends over time or other continuous variables.

```python
# Load the flights dataset
df = sns.load_dataset('flights')

# Create a line plot of the number of passengers over time
sns.lineplot(x='year', y='passengers', data=df)

# Show the plot
plt.show()
```

**Bar Plot**: Use `barplot()` function to create a bar plot. It is used to represent the distribution of a categorical variable.

```python
# Load the titanic dataset
df = sns.load_dataset('titanic')

# Create a bar plot of the survival rate by class
sns.barplot(x='class', y='survived', data=df)

# Show the plot
plt.show()
```

**Heatmap:** Use `heatmap()` function to create a heatmap. It is used to display data as color-encoded matrix.

```python
# Load the iris dataset
df = sns.load_dataset('iris')

# Create a heatmap of the correlation matrix
sns.heatmap(df.corr(), annot=True)

# Show the plot
plt.show()
```

**Box Plot:** Use `boxplot()` function to create a box plot. It is used to show the distribution of a numeric variable, highlighting any outliers.

```python
# Load the diamonds dataset
df = sns.load_dataset('diamonds')

# Create a box plot of the price by cut
sns.boxplot(x='cut', y='price', data=df)

# Show the plot
plt.show()
```
