# Intro to Decision Trees: A Hands-on Approach

A **Decision Tree** is a **Machine Learning** algorithm used for **classification** and **regression** tasks. It creates a tree-like structure where internal nodes represent a test on an attribute, branches represent the outcome of the test, and leaf nodes represent a class label. It partitions the data into subsets based on the values of the input features. The learning process is recursive, where the tree is constructed by recursively splitting the data into subsets. Decision trees are easy to interpret, handle categorical and numerical data, and can be visualized. But they can be prone to overfitting and can benefit from pruning.

The scikit-learn library in Python provides a simple and efficient tool for implementing decision tree algorithms. The `DecisionTreeClassifier` and `DecisionTreeRegressor` classes in the `sklearn.tree` module can be used for classification and regression tasks, respectively. These classes provide several parameters for controlling the tree's growth, such as `max_depth`, `min_samples_split`, and `min_samples_leaf`. The library also provides functions for visualizing decision trees, such as `plot_tree` and `export_graphviz`.

#### Classification

Here is an example of how to use the `DecisionTreeClassifier` class to train a decision tree on a dataset:

- Documentation: https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html

_Tip: It's a good practice to frequently check the documentation for the modules and classes you are using. The documentation provides detailed information about the available functions, methods, and parameters, and it can be very helpful in understanding how to use the library effectively._

First, the code imports the necessary modules and classes from scikit-learn: `DecisionTreeClassifier` and `train_test_split` from the `sklearn.tree` and `sklearn.model_selection` modules respectively.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

Next, the iris dataset is loaded using the `load_iris` function, which is a built-in dataset in scikit-learn. The dataset includes 150 samples of iris flowers, with four features (sepal length, sepal width, petal length, and petal width) and three classes (setosa, versicolor, and virginica). The features are stored in the `iris.data` variable, and the class labels are stored in the `iris.target` variable.

```python
iris = load_iris()
X = iris.data
y = iris.target
```

Scatter plot of the iris dataset using Seaborn:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Create a scatter plot of sepal length vs sepal width
sns.scatterplot(x="sepal_length", y="sepal_width", data=iris)

# Show the plot
plt.show()
```

The data is then split into training and test sets using the `train_test_split` function, which randomly splits the data into two groups. In this case, 80% of the data is used for training and 20% for testing. The function returns four variables: `X_train`, `X_test`, `y_train`, and `y_test`, which correspond to the feature vectors and class labels for the training and test sets, respectively.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

The decision tree classifier is then instantiated using the `DecisionTreeClassifier` class. The `random_state` parameter is set to 0, which is used to ensure reproducibility of the results.

```python
clf = DecisionTreeClassifier(random_state=0)
```

The classifier is trained on the training data using the `fit` method, which takes in the feature vectors and class labels of the training set as arguments.

```python
clf.fit(X_train, y_train)
```

Once the classifier is trained, it can be used to make predictions on new data using the `predict` method. In this case, the method is applied to the test set feature vectors, and the resulting predictions are stored in the `y_pred` variable.

```python
y_pred = clf.predict(X_test)
```

It's important to evaluate the classifier's performance using different metrics. Scikit-learn provides several built-in functions for computing different classification metrics such as accuracy, precision, recall, F1-score, and others.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Compute accuracy
acc = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}".format(acc))

# Compute precision
prec = precision_score(y_test, y_pred, average='weighted')
print("Precision: {:.2f}".format(prec))

# Compute recall
rec = recall_score(y_test, y_pred, average='weighted')
print("Recall: {:.2f}".format(rec))

# Compute F1-score
f1 = f1_score(y_test, y_pred, average='weighted')
print("F1-score: {:.2f}".format(f1))
```

#### Visualizing the decision tree

```python
from sklearn.tree import export_graphviz

print()

export_graphviz(clf, out_file='tree.dot', 
                feature_names=iris.columns[:4], 
                class_names=iris['species'].unique(), filled=True)
```
