# Imbalanced datasets

An imbalanced dataset is a dataset in which the classes are not represented equally. For example, if a dataset contains 95% of one class and only 5% of another class, it is considered imbalanced. This can be a problem in machine learning because models trained on imbalanced datasets may be biased towards the majority class, resulting in poor performance on the minority class. There are several techniques that can be used to address this issue, such as oversampling the minority class, undersampling the majority class, or using a combination of both. Additionally, using appropriate metrics such as precision, recall, F1-score, AUC-ROC, and G-mean can also help evaluate models performance in imbalanced datasets.

`imblearn` is a Python library for imbalanced-learn that provides a number of re-sampling techniques commonly used in machine learning to handle imbalanced datasets. It is built on top of popular machine learning libraries such as `numpy`, `scipy` and `scikit-learn`. imblearn provides a wide range of oversampling and undersampling methods, as well as ensemble and combination methods. Some of the most popular methods include `SMOTE`, `ADASYN`, `RandomUnderSampler`, `NearMiss`, and `SMOTE-ENN`. It also provides the ability to chain together different re-sampling techniques and machine learning models in a pipeline. With this library you can easily apply oversampling, undersampling, and combination techniques to your dataset to improve the performance of your machine learning models.

### Oversampling

```python
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt

# Generate an imbalanced dataset
X, y = make_classification(n_classes=2, class_sep=2,
                           weights=[0.1, 0.9], n_informative=3,
                           n_redundant=1, flip_y=0, n_features=20,
                           n_clusters_per_class=1, n_samples=1000,
                           random_state=10)

# Plot the class distribution before oversampling
plt.figure()
plt.title('Before oversampling')
plt.hist(y)

# Oversample the minority class using SMOTE
smote = SMOTE(random_state=12)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Plot the class distribution after oversampling
plt.figure()
plt.title('After oversampling')
plt.hist(y_resampled)
plt.show()
```

### Undersampling

```python
from sklearn.datasets import make_classification
from imblearn.under_sampling import RandomUnderSampler
import matplotlib.pyplot as plt

# Generate an imbalanced dataset
X, y = make_classification(n_classes=2, class_sep=2,
                           weights=[0.9, 0.1], n_informative=3,
                           n_redundant=1, flip_y=0, n_features=20,
                           n_clusters_per_class=1, n_samples=1000,
                           random_state=10)

# Plot the class distribution before undersampling
plt.figure()
plt.title('Before undersampling')
plt.hist(y)

# Undersample the majority class using RandomUnderSampler
rus = RandomUnderSampler(random_state=12)
X_resampled, y_resampled = rus.fit_resample(X, y)

# Plot the class distribution after undersampling
plt.figure()
plt.title('After undersampling')
plt.hist(y_resampled)
plt.show()
```
